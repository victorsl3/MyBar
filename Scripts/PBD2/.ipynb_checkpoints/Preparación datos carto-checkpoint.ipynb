```python
import pandas as pd
import datetime
```

```python
datos_camaras = pd.read_csv('datos_modelo_aforo.csv', sep=';').drop(['Unnamed: 0'], axis=1)
```

```python
datos_camaras.head()
```

```python
datos_camaras['FECHA']=datos_camaras['FECHA'].astype(str)
datos_camaras["FECHA"]
```

```python
datos_camaras['FECHA'] = datos_camaras['FECHA'].str.replace("-","/")
datos_camaras['FECHA']= datos_camaras['FECHA'].str.slice(0, 10)
datos_camaras["FECHA"]
```

```python
datos_camaras['HORA']=datos_camaras['HORA'].astype(str)
```

```python
datos_camaras['HORA']
```

```python
datos_camaras['FECHA']=datos_camaras['FECHA']+"T"+datos_camaras['HORA']+':00Z+0000'
```

```python
datos_camaras['FECHA']
```

```python
datos_camaras['FECHA'] = datos_camaras['FECHA'].str.replace("/","-")
```

```python
datos_camaras.head()
```

```python
datos_camaras.to_csv("datos_carto.csv",sep=";")
```
